The code for part 1.1 is given below 



```python
def get_data():
    xg = pd.read_csv('./xga.csv', delimiter=',')
    fouls = pd.read_csv('./fouls.csv', delimiter=',')
    xg = xg[['Result', 'xGA', 'Poss']]
    fouls = fouls[["Won%"]]
    merged = pd.concat([xg, fouls], axis=1, sort=False)
    merged = merged[merged["Result"]!='D']
    merged['Result'].replace("", np.nan, inplace=True)
    merged['xGA'].replace("", np.nan, inplace=True)
    merged['Won%'].replace("", np.nan, inplace=True)
    merged['Poss'].replace("", np.nan, inplace=True)
    merged.dropna(inplace=True)
    return merged
```

# Part 1.3

Since there are only 3 features, I decided to assume multivariate gaussian distribution conditioned on the 2 classes to make the prediction. This means that $P(X|y_i) \sim N(\mu, \sum) $ where $\mu$ is the mean of the data belonging to $y_i$ and $\sum$ is the covariance. 

So first, using the training data, I calculated the mean and covariance for the 2 mulitvariate gaussians, one for each of the 2 classes. And then I used Bayes theorem to make prediction for a test sample. 

$P(y|X_{test} = argmax_{y} (log(P(y)) + log(P(X_{test}|y)$)

To calculate $P(y)$, I just calculated the number of samples in each of the 2 classes. The code below is used to make prediction on the test data

```python
import numpy as np
import math
```

```python

def get_mean_and_cov():
    # get samples belonging to each of the 2 classes to get the class conditional mean and covariance of the multivariate gaussian distribution
    X_train_win = []
    X_train_loss = []
    for i in range(len(y_train)):
        if y_train[i] == 1:
            X_train_win.append(X_train[i])
        else: 
            X_train_loss.append(X_train[i])
    X_train_win = np.array(X_train_win)
    X_train_loss = np.array(X_train_loss)
    # getting the mean and cov
    cov_win = np.cov(X_train_win.T)
    mean_win = np.mean(X_train_win, axis = 0).reshape(-1,1)
    cov_loss = np.cov(X_train_loss.T)
    mean_loss = np.mean(X_train_loss, axis = 0).reshape(-1,1)
    return cov_win, cov_loss, mean_win, mean_loss
```

```python
cov_win, cov_loss, mean_win, mean_loss = get_mean_and_cov()
```

```python
# defining the formula for getting the log of the pdf
def get_log_pdf(mean, cov, X):
    a = -1/2 * math.log(np.linalg.det(cov))
    c = np.linalg.inv(cov)
    b = (X-mean).T@ c @ (X-mean)
    return  a -1/2 * (b[0][0])
```

```python
# getting the prior
prob_win = sum(y_train)/len(y_train)
prob_loss = 1 - prob_win
```

Once we have all the data needed to make the predictions like the prior and the mean and covariance of the gaussian, we are ready to make the predictions. These are made below

```python
def get_prediction(X):
    pdf_win = math.log(prob_win) + get_log_pdf(mean_win, cov_win, X)
    pdf_loss = math.log(prob_loss) + get_log_pdf(mean_loss, cov_loss, X)
    if pdf_win > pdf_loss:
        return 1
    return 0
```

```python
predictions = []
for i in range(len(X_test)):
    predictions.append(get_prediction(X_test[i].reshape(-1,1)))
    
```

Now we find the accuracy of our model

```python
correct = 0
for i in range(len(predictions)):
    if predictions[i] == y_test[i]:
        correct+=1
print(f'The accuracy of our model is {correct/len(y_test)*100}')
```

```python
x_win = []
y_win = []
x_loss = []
y_loss = []
for sample in np.random.multivariate_normal(mean, cov, 1000):
    prediction = get_prediction(sample.reshape(-1,1))
    if prediction:
        x_win.append(sample[0])
        y_win.append(sample[2])
    else:
        x_loss.append(sample[0])
        y_loss.append(sample[2])
plt.plot(x_loss, y_loss, linestyle='None', marker='o')
plt.plot(x_win, y_win, linestyle='None', marker='x')
plt.ylabel('percentage of tackles won')
plt.xlabel('expected number of goals Against')
plt.legend(["Loss", "Win"])
plt.show()    
```

# Introduction


Human Pose estimation is a computer vision task where we try to infer the pose of a person given an image or video. It is a key step towards understanding what a person is doing in a given image or video. Typically, to estimate the pose, we make try to identify and locate keypoints on the person in the image. For humans, these keypoints would typically be major joints which play the most important role in determining the posture.


The goal thus, is to try to identify and predict these keypoints in the given image/ video.


Almost the same technique can be used to estimate the pose for unanimated objects as well. One more thing to keep in mind is that the we can either do 2d pose estimation or 3d pose estimation. 2D pose estimation simply estimates the location of keypoints in 2D space relative to an image or video frame. The model estimates an X and Y coordinate for each keypoint. 3D pose estimation works to transform an object in a 2D image into a 3D object by adding a z-dimension to the prediction.


3D pose estimation allows us to predict the actual spatial positioning of a depicted person or object. It is much more difficult task than 2d pose estimation.


# Applications


Pose estimations has a huge number of applications. It can have a lot of applications in security as with this technique, automatically tracking human movement becomes possible. Pose estimation can also help us in activity recognition. For example, it can be used to automatically call ambulance when the model recognises that the person has fallen down. Variations of this technique can also be used to recognise gesture and gait which according to various studies, is also a unique fingerprint for a person.


Another application of this technique is for CGI applications. By tracking the variations of the human pose, the rendered graphics can be made to "naturally fit" the person as they move.


Finally, pose estimation is already being used for motion tracking in consoles. Kinect has been using 3D pose estimation to track the motion of the human players and to use it to render the actions of the virtual characters.


# Methodology 


Just like almost every other computer vision problem, deep learning architectures are used to solve this problem. In general, there are 2 overarching approaches: 


1)	Bottom up: In this method, the model detects every instance of a particular keypoint (eg all keypoints for the left arm) and then attempts to assemble groups of the keypoints into skeletons for distinct objects. 


2)	Top down: Inverse of the previous, the network first uses an object detector to draw a box around the different objects (like arm, leg etc) and then estimates the keypoints in each of these objects. 


Both these approaches contain deep learning models. Most of these models begin with an encoder which accepts an image as input and extracts features using a series of narrowing convolution blocks. What comes after depends on the method of pose estimation. The simplest method uses a regressor to output final predictions of each keypoint as a X, Y (and z for 3d) coordinate. However, empirically, this simple method doesn’t seem to perform as well. 


Second method uses an encoder decoder approach.  The output of the encoder is fed into a decoder. As an intermediate step, it generates heatmaps representing the likelihood that a keypoint is found in a given region of the image. After this, the exact coordinates of the keypoints are found by selecting heatmap locations with the highest keypoint likelihood. 


For top-down approaches, an object detection model is placed in between the encoder and the decoder and is used to crop regions of an image likely to contain an object. Keypoint heatmaps are then predicted individually for each box. Rather than having a single heatmap containing the likely location of all left hands in an image, we get a series of bounding boxes that should only contain a single keypoint of each type. This approach makes it easy to assign keypoints to specific instances without a lot of post-processing.


# Current good models


Now, I give a brief overview of some of the most popular deep learning models for this problem. 


1)	Stacked-Hourglass networks, Mask-RCNN, and other encoder-decoder networks
Pure encoder-decoder networks take an image as input and output heatmaps for each keypoint. They are a good place to start if you’re new to neural pose estimation. If you need to detect multiple poses, Mask-RCNN is a versatile architecture that predicts bounding boxes for objects in an image and then predicts poses within the regions of the image enclosed in the box.


2)	PersonLab / PoseNet and OpenPose
OpenPose and PersonLab (also known as PoseNet) are variants of an encoder-decoder architecture with a twist. In addition to outputting heatmaps, the model also outputs refinements to heatmaps in the form of short, mid, and long-range offsets. Heatmaps identify broad regions of an image where a keypoint is likely to be found, while offsets guide out predictions within a region to a more accurate final prediction


3)	Convolutional Pose Machines
Convolutional pose machines build on the encoder-decoder architecture by iteratively refining heatmap predictions using additional network layers and feature extraction. The final output is a single set of heatmaps, and post-processing involves identifying the pixels at which the heatmap probability is the highest for each keypoint.
