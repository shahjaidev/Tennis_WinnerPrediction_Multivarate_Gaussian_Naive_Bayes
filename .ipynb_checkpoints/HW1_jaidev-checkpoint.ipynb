# Models and approaches

RCNN- Region based Convolutional Neural Networks. 

Since there can be many objects in an image, to overcome the problem of selecting a huge number of regions, Ross Girshick [et al.CVPR14] proposed a method where we use Selective Search Algorithm to extract =~2000 Region Proposals.






Source: Fast R-CNN slides from ICCV15, Ross Girschik


To improve efficiency over doing a sliding window matrix loop over the original MxN image, state of the art RCNN object detection models first use a detector to generate region proposals. Some of the algorithms used to generate region proposals include After this, the CNN classifies the resized regions to the correct object. The bounding boxes around the detected object are subsequently refined by a support vector machine (SVM) that is trained using CNN features.

**Faster R-CNN:

Shaoqing Ren et al. came up with an object detection algorithm that eliminates the selective search algorithm and passes in feature map of the image into a RPN (region proposal network), letting it learn the region proposals and spit out regions of interest. 



Steps for a Faster R-CNN:
i) Region proposal network is applied on the feature maps, aka using a CNN as a feature extractor. This step returns the object proposals along with scores.
ii) RoI pooling layer is applied on these proposals to reduce all the proposals to the same size.

iii)These proposal regions are passed to a fully connected layer that has a softmax and a linear regression layer, which classifies the object and outputs bounding box parameters. 

In terms of speed, Faster R-CNN has a prediction time/image of 0.2 s compared to 40 s for a R-CNN. This is greatly important for real time self-driving car object detection like pedestrians. 

**Single Shot Detectors (SSD)

SSDs rely on a set of predetermined regions. A grid of anchor points is laid over the input image, and at each anchor point, boxes of multiple shapes and sizes serve as regions. Non maximum supression is a simple yet critical component of removing redundant bounding boxes and achieving usable detection results. We find detection bounding box posessing the highest confidence first and repeatedly eliminate all other boxes which have Intersection over Union metric over a pre-determined threshold with the chosen bounding box.

**YoloV3

You only look once, or YOLO, makes the tradeoff between speed and accuracy- providing very good, but not state of the art results in exchange for near real-time detection, something that is critical in self-driving car systems. 





The dimensions of the bounding box are predicted by applying a log-space transformation to the output vector of coordinates of the network.
